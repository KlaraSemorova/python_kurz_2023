```python
import pandas
import matplotlib.pyplot as plt
```

```python
temperatures = pandas.read_csv("temperature.csv")
print(temperatures)
```

```python
Helsinki = temperatures[temperatures["City"] == "Helsinki"]
city_Helsinki = pandas.DataFrame(Helsinki["AvgTemperature"], )
city_Helsinki = city_Helsinki.reset_index(drop=True)

Miami_Beach = temperatures[temperatures["City"] == "Miami Beach"]
city_Miami_Beach = pandas.DataFrame(Miami_Beach["AvgTemperature"])
city_Miami_Beach= city_Miami_Beach.reset_index(drop=True)

Tokyo = temperatures[temperatures["City"] == "Tokyo"]
city_Tokyo = pandas.DataFrame(Tokyo["AvgTemperature"])
city_Tokyo = city_Tokyo.reset_index(drop=True)
```

```python
cities = pandas.DataFrame(columns = ["Helsinky", "Miami Beach", "Tokyo"])
cities["Helsinky"] = city_Helsinki
cities["Miami Beach"] = city_Miami_Beach
cities["Tokyo"] = city_Tokyo
cities
```

```python
cities.plot(kind='box', whis=[0, 100])
# Teploty jsou v listopadu nejnižší v Helsinkách a nejvyšší v Miami Beach.
```
