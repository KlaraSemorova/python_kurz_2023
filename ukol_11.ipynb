```python
# 2. Vymysli regulární výraz, který označí všechny "poslední řádky adresy" v textu. Poslední řádka adresy zpravidla obsahuje PSČ a název obce, například 190 16 PRAHA 916 nebo 742 45 FULNEK. Celkem by jich mělo být 18.

# regex výsledek: \d{3} ?\d{2} {1,2}.+
```

```python
# bonus
import re

uzivatelske_jmeno = re.compile(r"[a-zA-Z]{6,10}")
zadane_uzivatelske_jmeno = input("Přihlašovací jméno: ")
kontrola_uzivatelske_jmeno = uzivatelske_jmeno.fullmatch(zadane_uzivatelske_jmeno)

if kontrola_uzivatelske_jmeno:
    email = re.compile(r"[\w\.\-+\"]{1,64}@[\w\.\-\"]{1,253}\.[1-9a-z]{1,253}$")
    zadany_email = input("E-mailová adresa: ")
    kontrola_email = email.fullmatch(zadany_email)
    if kontrola_email:
        heslo = re.compile(r"[\w_\-\.+=]{12,30}")
        zadane_heslo = input("Heslo: ")
        kontrola_heslo = heslo.fullmatch(zadane_heslo)
        if kontrola_heslo:
            print("Zadané heslo je v pořádku. Děkujeme.")
        else:
            print("Zadali jste neplatné heslo.")
    else:
        print("Zadali jste neplatný email.")
else:
    print("Zadali jste neplatné uživatské jméno.")

```
